```python
import pandas as pd
```

```python
fornecedor1 = pd.read_csv('temp/fornecedor1.csv', sep = ';', encoding= 'utf8', dtype=object)
fornecedor1
```

```python
fornecedor2 = pd.read_csv('temp/fornecedor2.csv', sep = ';', encoding= 'utf8', dtype=object)
fornecedor2
```

```python
def newRows(df1, df2, key):

    df3 = pd.merge(df1.sort_values([key]),df2.sort_values([key]), on=key, how='right', indicator = True)
    
    df3 = df3[df3['_merge'] == 'right_only']

    ## Seleciona a coluna pos filtro com os indices criados para tabela da direita, referênciados por _y.
    df3 = df3[(col for col in df3.columns if col[-1] == 'y' or col == key)]

    return df3
```

```python
newRows(fornecedor1, fornecedor2, 'cod_fornecedor')
```
